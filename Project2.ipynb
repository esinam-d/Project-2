```python
import pandas as pd
import matplotlib.pyplot as plt

file_name = 'james_bond_data_cleansed.csv'
james_bond_data = pd.read_csv(file_name)

print(james_bond_data.head())

#To create the scatter plot for question 1: Scatter plot of IMDb vs Rotten Tomatoes ratings
plt.figure(figsize=(10,6))

plt.scatter(james_bond_data['imdb'], james_bond_data['rotten_tomatoes'], color='blue',alpha=0.7)
plt.xlabel('Average IMDb Rating')
plt.ylabel('Average Rotten Tomatoes Rating')
plt.title('Scatter Plot Ratings')

#There is a positive correlation
plt.show()


#To create the scatter plot for question 2: Scatter plot of IMDb vs Bond kills
plt.figure(figsize=(10,6))

plt.scatter(james_bond_data['imdb'], james_bond_data['bond_kills'], color='blue',alpha=0.7)
plt.xlabel('IMDb Rating')
plt.ylabel('Bond Kills')
plt.title('Bond kills vs IMDb Rating')

#There is no correlation or relationship between IMDb rating and the number of bad guys killed by Bond
plt.show()

#To create the legend for question 3
plt.figure(figsize=(10,6))

plt.plot(james_bond_data['release_year'], james_bond_data['movie_budget'], label='movie budget', color='blue', marker='o')

plt.plot(james_bond_data['release_year'], james_bond_data['income_world'], label='income world', color='red', marker='o')

plt.xlabel('Release Year')
plt.ylabel('Amount(in Billion USD)')

plt.legend()

plt.grid(alpha=0.5)
#There is an upward trend which means higher revenues.
plt.show()

# To create the bar plot for question 4
plt.figure(figsize=(10,6))

james_bond_data['income_budget_ratio'] =james_bond_data['income_world']/james_bond_data['movie_budget']

plt.figure(figsize= (10,5))
plt.bar(james_bond_data['release_year'],james_bond_data['income_budget_ratio'], color= 'blue', alpha=0.7)

plt.xlabel('Release Year')
plt.ylabel('Income World-Budget Ratio')
plt.title('Income World-Budget ratio along the time')

# The trend of income world – movie budget ratio along the time? 

plt.show()
```
